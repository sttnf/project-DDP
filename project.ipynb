```python
import json

# Data awal
data = {
    "1": {"Kode": "ABCD1234", "Nama Barang": "Kecap Bebek 150 ml", "Harga": 15000, "Jml": 3, "Total": 45000},
    "2": {"Kode": "ABCD1235", "Nama Barang": "Plastik Sampah", "Harga": 500, "Jml": 20, "Total": 10000},
    # Data lainnya
}

def login():
    print("Selamat datang!")
    username = input("Username: ")
    password = input("Password: ")

    if username == "admin" and password == "admin123":
        return "admin"
    elif username == "user" and password == "user123":
        return "user"
    else:
        print("Login gagal. Silakan coba lagi.")
        return None

def show_products(filter_by=None, sort_by=None, search_query=None):
    products = list(data.values())

    if filter_by == "nama-asc":
        products.sort(key=lambda x: x["Nama Barang"])
    elif filter_by == "nama-desc":
        products.sort(key=lambda x: x["Nama Barang"], reverse=True)
    elif filter_by == "harga-asc":
        products.sort(key=lambda x: x["Harga"])
    elif filter_by == "harga-desc":
        products.sort(key=lambda x: x["Harga"], reverse=True)

    if search_query:
        products = [p for p in products if search_query.lower() in p["Nama Barang"].lower()]

    for product in products:
        print(f"Kode: {product['Kode']}, Nama: {product['Nama Barang']}, Harga: {product['Harga']}, Stok: {product['Jml']}")

def add_product():
    print("Menambahkan Produk Baru:")
    kode = input("Kode Produk: ")
    nama = input("Nama Produk: ")
    harga = int(input("Harga Produk: "))
    jumlah = int(input("Jumlah Stok: "))
    total = harga * jumlah

    new_product = {
        "Kode": kode,
        "Nama Barang": nama,
        "Harga": harga,
        "Jml": jumlah,
        "Total": total
    }
    data[str(len(data) + 1)] = new_product
    print("Produk berhasil ditambahkan.")

def buy_product():
    print("Membeli Produk:")
    kode_produk = input("Masukkan kode produk: ")
    jumlah = int(input("Masukkan jumlah: "))

    for item in data.values():
        if item["Kode"] == kode_produk:
            if item["Jml"] >= jumlah:
                total = item["Harga"] * jumlah
                print(f"Anda membeli {jumlah} {item['Nama Barang']} seharga {total}.")
                item["Jml"] -= jumlah
                item["Total"] = item["Harga"] * item["Jml"]
                print("Pembelian berhasil.")
                return
            else:
                print("Stok tidak mencukupi.")
                return
    print("Kode produk tidak ditemukan.")

def show_menu():
    print("Menu:")
    print("1. Tampilkan Produk")
    print("2. Tambah Produk")
    print("3. Beli Produk")
    print("4. Keluar")

def run_app():
    role = login()
    if role == "admin":
        while True:
            show_menu()
            choice = input("Masukkan pilihan Anda: ")
            if choice == "1":
                print("\nFilter Produk:")
                print("1. Urutkan A-Z")
                print("2. Urutkan Z-A")
                print("3. Urutkan Harga Termurah")
                print("4. Urutkan Harga Termahal")
                print("5. Cari Produk")
                filter_choice = input("Masukkan pilihan filter: ")
                if filter_choice == "1":
                    show_products(filter_by="nama-asc")
                elif filter_choice == "2":
                    show_products(filter_by="nama-desc")
                elif filter_choice == "3":
                    show_products(filter_by="harga-asc")
                elif filter_choice == "4":
                    show_products(filter_by="harga-desc")
                elif filter_choice == "5":
                    search_query = input("Masukkan kata kunci pencarian: ")
                    show_products(search_query=search_query)
                else:
                    print("Pilihan tidak valid.")
            elif choice == "2":
                add_product()
            elif choice == "3":
                print("Maaf, fitur ini hanya untuk user.")
            elif choice == "4":
                print("Keluar dari aplikasi.")
                break
            else:
                print("Pilihan tidak valid.")
    elif role == "user":
        while True:
            show_menu()
            choice = input("Masukkan pilihan Anda: ")
            if choice == "1":
                print("\nFilter Produk:")
                print("1. Urutkan A-Z")
                print("2. Urutkan Z-A")
                print("3. Urutkan Harga Termurah")
                print("4. Urutkan Harga Termahal")
                print("5. Cari Produk")
                filter_choice = input("Masukkan pilihan filter: ")
                if filter_choice == "1":
                    show_products(filter_by="nama-asc")
                elif filter_choice == "2":
                    show_products(filter_by="nama-desc")
                elif filter_choice == "3":
                    show_products(filter_by="harga-asc")
                elif filter_choice == "4":
                    show_products(filter_by="harga-desc")
                elif filter_choice == "5":
                    search_query = input("Masukkan kata kunci pencarian: ")
                    show_products(search_query=search_query)
                else:
                    print("Pilihan tidak valid.")
            elif choice == "2":
                print("Maaf, fitur ini hanya untuk admin.")
            elif choice == "3":
                buy_product()
            elif choice == "4":
                print("Keluar dari aplikasi.")
                break
            else:
                print("Pilihan tidak valid.")

run_app()
```
